```python
# How do property prices vary by property type?

avg_price = properties.groupby("Type")["Listing_Price"].mean().round(2).sort_values(ascending=False).reset_index(name="Avg_Listing_Price")
avg_price
```

```python
# Which Property Types Are Most Frequently Listed?

listing_count= properties["Type"].value_counts().sort_values(ascending=False)
listing_count
```

```python
# Which property types are most frequently listed in each city?

properties.groupby(["Type","City"]).size().reset_index(name="Total_Listings").sort_values(by="Total_Listings",ascending=False)
```

```python
# Which Cities Have the Highest Average Home Prices?

avg_home_price = market.groupby("City")["Avg_Home_Price"].mean().sort_values(ascending=False).round(2).reset_index(name="Avg_Home_Price")
avg_home_price
```

```python
 # How Have Home Prices Changed Over Time?

# Step 1: Calculate average home price for each city and year

avg_price_year = (
    market.groupby(["City", "Year"])["Avg_Home_Price"]
    .mean()
    .reset_index()
)

# Step 2: Get previous year's average home price for each city

avg_price_year["Previous_Price"] = (
    avg_price_year.groupby("City")["Avg_Home_Price"]
    .shift(1)
)

# Step 3: Calculate Year-over-Year (YoY) Growth Percentage

avg_price_year["YoY_Growth (%)"] = (
    (avg_price_year["Avg_Home_Price"] - avg_price_year["Previous_Price"])
    / avg_price_year["Previous_Price"]
) * 100

# Round the growth values

avg_price_year["YoY_Growth (%)"] = avg_price_year["YoY_Growth (%)"].round(2)

# final result

avg_price_year
```

```python
# Which City Recorded the Highest Property Values Each Year?

highest_price = market.sort_values(by="Avg_Home_Price",ascending=False).groupby("Year")[["City","Avg_Home_Price"]].first().reset_index()
highest_price["Avg_Home_Price"] = highest_price["Avg_Home_Price"].round(2)
highest_price
```

```python
# Which City Recorded the Lowest Property Values Each Year?
lowest_price = market.sort_values(by="Avg_Home_Price").groupby("Year")[["City","Avg_Home_Price"]].first().reset_index()
lowest_price["Avg_Home_Price"]= lowest_price["Avg_Home_Price"].round(2)
lowest_price
```

```python
# Which Cities Exhibit the Strongest Housing Demand, and What Are the Typical Home and Rental Prices?

housing_demand = (market.groupby("City").agg({"Housing_Demand_Index":"mean", "Avg_Home_Price":"mean","Avg_Rent_Price":"mean"})
                    .round(2)
                    .sort_values(by="Housing_Demand_Index",ascending=False)
                    .reset_index())
housing_demand
```

```python
# Which Cities Attract the Highest Investor Activity?

investor_activity = (market.groupby("City")["Investor_Activity_Score"].mean().round(2)
    .sort_values(ascending=False).reset_index(name="Avg_Investor_Activity"))

investor_activity["Investor_Rank"] = investor_activity["Avg_Investor_Activity"].rank(method="dense",ascending=False)
investor_activity
```

```python
# Which Income Groups Face the Greatest Housing Affordability Challenges?

affordability = (
    market.groupby("Income_Bracket")["Affordability_Price_to_Income_Ratio"]
    .mean()
    .round(2)
    .reset_index(name="Price_to_Income_Ratio")
)

def affordability_status(ratio):
    if ratio >= 6:
        return "Severely Unaffordable"
    elif ratio >= 5:
        return "Moderately Unaffordable"
    elif ratio >= 4:
        return "Slightly Unaffordable"
    else:
        return "Affordable"

affordability["Affordability_Stress"] = (
    affordability["Price_to_Income_Ratio"]
    .apply(affordability_status)
)

affordability
```

```python
#Q1. Which Years Recorded the Highest Number of New Property Developments?

new_construction = market.groupby("Year")["New_Construction_Count"].sum().sort_values(ascending=False).reset_index(name="New_Properties_built")
new_construction
```

```python
# Which Cities Experienced the Highest Level of New Construction Activity?
new_construction_city = (
    market.groupby("City")["New_Construction_Count"]
    .sum()
    .reset_index(name="New_Properties_Built")
    .sort_values(by="New_Properties_Built", ascending=False)
)

new_construction_city
```

```python
# How Does New Construction Influence Home Prices?
construction_price = (
    market.groupby("Year")
    .agg({
        "New_Construction_Count": "sum",
        "Avg_Home_Price": "mean"
    })
    .round(2)
    .reset_index()
)

construction_price
```

```python
# How Do Interest Rates Impact Average Home Prices?

interest_rate = market.groupby("Interest_Rate")["Avg_Home_Price"].mean().round(2).reset_index(name="Average_Home_Price")
interest_rate["Previous_Price"]= interest_rate["Average_Home_Price"].shift(1)
interest_rate["Percent_change"] = (interest_rate["Average_Home_Price"]-interest_rate["Previous_Price"])/interest_rate["Previous_Price"]*100
interest_rate["Percent_change"]=interest_rate["Percent_change"].round(2)
interest_rate
```
